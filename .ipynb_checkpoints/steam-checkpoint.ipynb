```python
import pandas as pd
```

```python
csv_file = "./steam_games.csv"
xlsx_file_spy = "./SteamSpy.xlsx"
```

```python
steam_data_df = pd.read_csv(csv_file)
steam_data_df.head()
```

```python
steam_spy_df = pd.read_excel(xlsx_file_spy)
steam_spy_df.head()
```

```python
new_steam_data_df = steam_data_df[['name',
                                   'all_reviews',
                                   'release_date',
                                   'developer',
                                   'genre',
                                   'original_price']].copy()
new_steam_data_df.head()
```

```python
new_steam_spy_df = steam_spy_df[["Game", "Owners"]].copy()
new_steam_spy_df.head()
```

```python
new_steam_data_df.rename(columns={'original_price': 'Price',
                                  'name' : 'Game',
                                  'all_reviews' : 'reviews',
                                  'release_date' : 'release date'},inplace=True)
new_steam_data_df
```

```python
merged_steam_df = pd.merge(new_steam_data_df, new_steam_spy_df, on="Game")
pd.set_option('display.max_rows', 300)
merged_steam_df
```

```python
#pd.reset_option('max_columes')
```

```python
x = 0
for index, row in merged_steam_df.iterrows():
    try:
        review_split = row["reviews"]
    
        percent_split1 = review_split.split(",", 1)
        percent_split2 = percent_split1[1].split("of", 1)
        percent_split3 = percent_split2[0].split("-",1)
        print(x,percent_split3[1])
        x += 1
        merged_steam_df.loc[index, 'reviews'] = percent_split3[1]
    except:
        continue
```

```python
merged_steam_df
```

```python
merged_steam_df = merged_steam_df.dropna()
merged_steam_df
```

```python
merged_steam_df['Price'].mask(merged_steam_df['Price'] == 'Free', '$0.00', inplace=True)
merged_steam_df['Price'].mask(merged_steam_df['Price'] == 'Free to Play', '$0.00', inplace=True)
merged_steam_df['Price'].mask(merged_steam_df['Price'] == 'Free To Play', '$0.00', inplace=True)
merged_steam_df['Price'].mask(merged_steam_df['Price'] == 'SpeedRunners Party Mode', '$0.00', inplace=True)
merged_steam_df['Price'].mask(merged_steam_df['Price'] == 'Soundpad Demo', '$0.00', inplace=True)
merged_steam_df['Price'].mask(merged_steam_df['Price'] == 'LudoNarraCon Demo', '$0.00', inplace=True)
merged_steam_df
```

```python
merged_steam_df['Price'] = merged_steam_df['Price'].replace({'\$': ''}, regex=True).astype(float)
merged_steam_df['release date'] = pd.to_datetime(merged_steam_df['release date'])
#merged_steam_df['Owners'].str.replace('\.\.','-')
merged_steam_df['Owners'] = merged_steam_df['Owners'].str.replace('\.\.','-')
#merged_steam_df['genre'] = pd.tolist(merged_steam_df['genre'])
#merged_steam_df = merged_steam_df['genre'].to_list()
merged_steam_df
```
